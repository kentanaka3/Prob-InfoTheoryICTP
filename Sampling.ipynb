```python
# Fundamental imports
import numpy as np
import matplotlib.pyplot as plt
```

$$P\{k| n,p\}=\left(\begin{matrix}n\\ k\end{matrix}\right)p^k(1-p)^{n-k}$$

```python
p = 0.4
N = np.random.uniform(0.0, 1.0, 1000) < p
plt.plot([sum(N[:i + 1])/(i + 1) for i, k in enumerate(N)], label="Sampling")
plt.plot([p]*len(N), label=f"$p = {p}$")
plt.xlabel("N")
plt.ylabel("k/N")
plt.legend()
plt.show()
```

$$P\{k_{n + 1} = k_n + 1| k_n = k, n\}=\frac{k + a}{n + 2a}$$

```python
a = 1.0
elem = 1000
N = 2*a*np.arange(1, elem + 1)*np.random.uniform(0.0, 1.0, elem)

def rcv(N):
  if len(N) > 1:
    N[:-1] = rcv(N[:-1])
    N[-1] = N[-2] + int(N[-1] < N[-2] + a)
    return N
  else:
    return [int(N[0] < a)]

plt.loglog()
plt.plot([k/(i + 1) for i, k in enumerate(rcv(N))], label=f"a = {a}")

plt.xlabel("N")
plt.ylabel("k/N")
plt.legend()
```
